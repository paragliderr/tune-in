```python
import requests
from datetime import datetime, timezone

# TOKEN
ACCESS_TOKEN = "64c49626a949a163cc274b2ab194118b8dd51bde"  # from your refresh script
code = "557d8f797c23cf97a8134cddf64b5f3f690c2b3f"

headers = {
    "Authorization": f"Bearer {ACCESS_TOKEN}"
}

# FETCH ALL ACTIVITIES
activities = []
page = 1

while True:
    url = f"https://www.strava.com/api/v3/athlete/activities?page={page}&per_page=30"

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print("API Error:", response.text)
        break

    data = response.json()

    if not data:  # no more data
        break

    activities.extend(data)
    page += 1

# today
today = datetime.now(timezone.utc).date()

total_distance = 0      # meters
total_time = 0          # seconds

for activity in activities:
    activity_date = datetime.strptime(
        activity["start_date"],
        "%Y-%m-%dT%H:%M:%SZ"
    ).date()

    if activity_date == today and activity["type"] == "Run":
        total_distance += activity["distance"]
        total_time += activity["moving_time"]


distance_km = total_distance / 1000
time_min = total_time / 60

if distance_km > 0:
    pace = time_min / distance_km   # min/km
else:
    pace = 0


print("TODAY")
print("Distance (km):", round(distance_km, 2))
print("Time (min):", round(time_min, 2))
print("Avg Pace (min/km):", round(pace, 2))

#  processed
processed_data = {
    "date": str(today),
    "distance_km": round(distance_km, 2),
    "time_min": round(time_min, 2),
    "avg_pace": round(pace, 2),
    "activity_type": "run"
}
print("\nStructured Data:", processed_data)
```
